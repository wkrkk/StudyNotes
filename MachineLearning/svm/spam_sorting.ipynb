### 垃圾邮件分类

使用SVM来构建垃圾邮件过滤器

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.io import loadmat
```

```python
spam_train = loadmat('./datasets/spamTrain.mat')
spam_test = loadmat('./datasets/spamTest.mat')

spam_train, spam_test
```

```python
X = spam_train['X']
Xtest = spam_test['Xtest']
y = spam_train['y'].ravel()
ytest = spam_test['ytest'].ravel()

X.shape, y.shape, Xtest.shape, ytest.shape
```

```python
from sklearn import svm
# 分类器
svc = svm.SVC()
svc.fit(X, y)
```

```python
# np.round() 对给定的数组进行四舍五入，能够指定精度
print(svc.score(X, y))
print('Training accuracy = {0}%'.format(np.round(svc.score(X, y) * 100, 2)))
print('Test accuracy = {0}%'.format(np.round(svc.score(Xtest, ytest) * 100, 2)))
```

可视化结果

```python
# 生成对角矩阵,主对角线全1，其余全0
kw = np.eye(1899)
print(kw)
```

```python
kw[:3,:]
```

```python
spam_val = pd.DataFrame({'idx':range(1899)})
print(spam_val)
```

```python
spam_val['isspam'] = svc.decision_function(kw)
```
